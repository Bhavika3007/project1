```python
train=pd.read_csv("C:/Users/CVR/Downloads/titanic/train.csv")
```

```python
train
```

```python
sns.barplot(data=train,x='Pclass',y='Survived',palette='viridis',errorbar=None)
plt.title('Survived rate by passengers')
plt.xlabel('Passenger')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.scatterplot(data=train,x='Age',y='Fare',hue='Age',palette='magma',alpha=0.6)
plt.title('Relation between age and fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
pairwise_data = train[['Age', 'Fare', 'Pclass','Survived']].dropna()
sns.pairplot(pairwise_data, hue='Survived', palette='husl',diag_kind='kde')
plt.suptitle('Pairwise Relationship',y=1.02)
plt.show()
```

```python
num=train.select_dtypes(include='number')
corr_matrix=num.corr()
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
