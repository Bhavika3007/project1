```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.title("simple line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

plt.plot(x,y, color="blue",linestyle='dashed', linewidth=2, marker='o')
plt.show()
```

```python
x=['jan','feb','mar','apr','may','june']
y=[1000,2300,4500,1200,1500,300]
plt.plot(x,y, color='green',linestyle='dashed',marker='o')
plt.title("Growth of Savings Account")
plt.xlabel("Months")
plt.ylabel("Savings")
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values, color=['skyblue','pink'], edgecolor='black',width=0.75)
plt.title("Basic Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
ages=[20,22,12,23,25,27,28,30,31,32,14,35,35,38,67,20,12,33,44,50,13,11,7,89,10,3,5,45,46,89,67,23,44,44]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='skyblue',edgecolor='black')
plt.title("Age Distribution of Customer")
plt.title("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Python', 'Java', 'C', 'HTML', 'JavaScript']
values = [40, 15, 7, 35, 9]
colors = ['lightgreen', 'orange', 'red', 'violet', 'yellow']
plt.pie(values, labels=categories, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage")
plt.show()
```

```python
x=[5,10,15,20,25]
y=[10,20,25,10,35]
plt.scatter(x,y,color='green',s=100,edgecolor='black')
plt.title("scatter plot example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```
